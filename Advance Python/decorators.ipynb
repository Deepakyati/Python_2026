Decorators

```python
#Decorators

#modifies another function without changing its code
#like iceing on cake
#for creating a decorator you firsts have to create a decorator function and then inside that we will create a wrapper
#
```

```python
class Animal:
    def show(self):
        print('hello how are you ?')


obj=Animal()
obj.show()
```

```python
class Animal:
    @property
    def show(self):
        print('hello how are you ?')


obj=Animal()
obj.show #running without ()
```

```python
#implementation

def decorate(func): #func is function argument, can be anything
    def wrapper():
        print('I will print before the function hello')
        func()
        print('I will print after the function hello')
    return wrapper


@decorate #name can be anything
def hello():
    print('hello i am deepak')


hello()
```

```python
def decorate(func): #func is function argument, can be anything
    def wrapper(a,b):
        print('the sum of your numbers are')
        func(a,b)
        print('Thanks, hope you like it!')
    return wrapper


@decorate #name can be anything
def addition(a,b):
    print(f'total is {a+b}')


addition(2,3)
```

```python
def decorate(func): #func is function argument, can be anything
    def wrapper(a,b):
        print('the sum of your numbers are')
        func(a,b)
        print('Thanks, hope you like it!')
    return wrapper


@decorate #name can be anything
def addition(a,b,c):
    print(f'total is {a+b+c}')


addition(2,3,4)
```

```python
#decorstors with *args

def decorate(func): #func is function argument, can be anything
    def wrapper(*args):
        print('the sum of your numbers are')
        func(*args)
        print('Thanks, hope you like it!')
    return wrapper


@decorate #name can be anything
def addition(a,b,c):
    print(f'total is {a+b+c}')


addition(2,3,4)
```
