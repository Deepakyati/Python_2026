Lambda_functions

```python
#Lambda_functions.ipynb

#anonymous, inline function defined using the lambda keywords
#often used for short, simple function that are used only once or temporarily
#you can have multiple arguments but there will be only one experession
```

```python
def addition(a,b):
    print(a+b)

addition(4,5)
```

```python
square = lambda x:x**2
print(square(9))
```

```python
check_even = lambda x: x%2==0

check_even(11)
```

```python
addition =lambda a,b:a+b

addition(5,6)
```

```python
check_even_odd = lambda x: "Even" if x%2==0 else "Odd"

print(check_even_odd(11))
print(check_even_odd(12))
```
