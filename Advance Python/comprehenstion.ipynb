Comprehension

```python
#list comprehension

l=[i for i in range(1,21) if i%2==0]
l
```

```python
#print a list with cube numbers from 1-10
l1=[x**3 for x in range(1,11)]
l1
```

```python
#Dict comprehenstion, just use {}

l2={i:i**2 for i in range(1,100) if i%2==0}
l2
```
