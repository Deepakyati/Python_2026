*args & *kwargs

```python
#args and kwargs
```

```python
def add(a,b):
    print(a+b)


add(3,4)
```

```python
def add(a,b):
    print(a+b)


add(3,4,5)
```

```python
def add(*args):
    print(args)


add(3,4,5)
```

```python
def add(*args):
    print(args)


add(3,4,5,4,5,6,7,8,9)
```

```python
def add1(*args):
    sum=0
    for i in args:
        sum+=i
    print(sum)

add1(3,4,5,4,5,6,7,8,9)
```

```python
def add1(*s): # * should be there
    sum=0
    for i in s:
        sum+=i
    print(sum)

add1(3,4,5,4,5,6,7,8,9)
```

**kwargs

```python
#**kwargs


def add2(**kwargs):
    print(kwargs)


add2(a=12,b=13,c=14,d=15)
```

```python
def information(**kwargs):
    print('your information: ')
    for i in kwargs:
        print(f'{i}:{kwargs[i]}')
    


information(name='Deepak', age=30, designation='AI/ML')
```

```python
#decorators *args **kwargs


def decorate(func): #func is function argument, can be anything
    def wrapper(*args,**kwargs):
        print('the sum of your numbers are')
        func(*args,**kwargs)
        print('Thanks, hope you like it!')
    return wrapper


@decorate #name can be anything
def addition(a,b,c):
    print(f'total is {a+b+c}')


addition(2,3,4)
```
