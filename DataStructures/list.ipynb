```python
#Print positive and negative elementsof an list

l=[-10,-20,30,-40,50,-60,70,-11,-22,33,44,55]

print('positive elements are: ')
for i in l:
    if i>=0:
        print(i)
print('negative elements are: ')
for i in l:
    if i<0:
        print(i)
```

```python
#Mean of a list elements

l1=[10,20,30,40,50,60,70]

sum=0
for i in l1:
    sum+=i
    mean=sum/len(l1)
print(sum)
print(mean)
```

```python
#find greatest element and print its index too

l2=[10,20,30,40,120,50,60,70,15,65,95,150,100, 105, 130, 166]

greatest = 0
for i in range(len(l2)):
    if l2[i]>greatest:
        greatest = l2[i]

print(greatest)
print(l2.index(greatest))
```

```python
#find the second greatest element in a list

l3=[10,50,20,60,30,70,40, 99,97]

greatest=0
second_greatest =0

for i in range(len(l3)):
    if l3[i]>greatest:
        greatest = l3[i]

for i in range(len(l3)):
    if l3[i]>second_greatest and l3[i]<greatest:
        second_greatest = l3[i]

print(greatest)
print(second_greatest)
```

```python
#find the second greatest element in a list

l3=[10,50,20,60,30,70,40, 99,97]

greatest=l3[0]
second_greatest =l3[0]

for i in l3:
    if i>greatest:
        second_greatest = greatest
        greatest = i
    elif i>second_greatest:
        second_greatest =i

print(greatest, second_greatest)
```
