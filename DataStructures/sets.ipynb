```python
#sets Methods

s1={1,2,8,9,3,4,5,6,7}

s1.add(18)

print(s1)

s1.remove(8)
print(s1)

s1.discard(19) #if num is present it will remove, if not present then it will continue with no error
print(s1)

s1.pop()
print(s1)

s1.pop()
print(s1)

s1.pop()
print(s1)
```

```python
#union
#intersection
#difference
#symmetric


a={1,2,3,4,5}
b={4,5,6,7,8}

c=a.intersection(b)
print(c)

d=a.union(b)
print(d)

e=a|b #can use '|' symbol for union
print(e)


f=a&b #can use '&' symbol for intersection
print(f)

g=a.difference(b)
print(g)

h=b.difference(a)
print(h)


#symmetric difference
i=a.symmetric_difference(b)
print(i)

j=a^b #can use '^' symbol for symmetric diff
print(j)



```
