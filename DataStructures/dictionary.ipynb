Deep copy vs Shallow copy

```python
#deep copy vs shallow copy
#deep copy
a=[1,2,3,4,5]
b=a
b[0]=10

print(a)
print(b)
```

```python
#shallow copy
a=[1,2,3,4,5]
b=a.copy()
b[0]=10

print(a)
print(b)
```

Dict Questions

```python
#merge

d1={1:10,2:20,3:30}
d2={4:40,5:50,6:60}

for i in d2:
    d1[i]=d2[i]

print(d1)
```

```python
#sum of values
d1={1:10,2:20,3:30}
sum=0

for i in d1:
    sum=sum+d1[i]

print(sum)
```

```python
#count the frequency of each element in list

l=[1,1,1,1,1,2,2,2,3,3,4,5,5,6,6,7,7,8,9,9]

dict ={}

for i in l:
    if i in dict.keys():
        dict[i]+=1
    else:
        dict[i]=1

print(dict)
    
```

```python
#combine two dictionary

d1={1:10,2:20,3:30,4:45, 5:500}
d2={4:40,5:50,6:60}

for i in d2:
    if i in d1.keys():
        d1[i]+=d2[i]
    else:
        d1[i]=d2[i]

print(d1)
```
