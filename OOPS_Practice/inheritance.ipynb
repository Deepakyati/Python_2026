Inheritance

```python
#inheritance
#child inherit properties from parent
#works between classes

#code reusability
#organized structure
#easy tomaintain and extend
```

```python
class FactoryMumbai: #superclass/parent class
    a='i am an attribute inside Factory'
    def hello(self):
        print('i am a method inside Factory')


class FactoryHyderabad(FactoryMumbai): #subclass/child class
    pass

obj = FactoryMumbai()
obj2 = FactoryHyderabad()

print(obj.a)
print(obj2.a)
print(obj2.hello()) #obj2.hello() ->None

```

```python
class Animal:

    def __init__(self,name):
        self.name=name

    def show(self):
        print(f'hello your name is{self.name}')


class Human(Animal):
    def __init__(self,name,age):
        super().__init__(name)
        self.age=age

    def show(self):
        print(f'hello your name is{self.name}, {self.age}')

person1= Human(' deepak',32)
person1.name
```

```python
#Method Overriding

person1.show()
```

```python
class Factory:
    def __init__(self, material,zips):
        self.material = material
        self.zips = zips



class BhopalFactory:
    def __init__(self, material,zips,color):
        super().__init__(material,zips)
        self.color = color

class PuneFactory:
    def __init__(self, material,zips,color, pockets):
        super().__init__(material,zips,color)
        self.pockets = pockets


```
