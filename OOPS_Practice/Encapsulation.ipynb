Encapsulation

```python
#Encapsulation
#putting data and code together
#hiding internal details of class

#Keep data safe from being changed by mistake
#it makes your code clean and easy to use
#it gives control over what other can access or change , protected, Private access modifier
```

```python
class Factory:
    __a='Pune'

    def show(self):
        print('Hello i am in Pune factory')


obj=Factory()

obj.show()

obj.a='bhopal' 

print(obj.a)
```

```python
#access modifier in Python
#3 types

#Public - inherited class can acess
#Protected: _a Python doesn't enforce this, but uses a naming convetion to tell developer
#Private: __a
```

```python
class Factory:
    __a='Pune'

    def show(self):
        print('Hello i am in Pune factory')

class Mumbai(Factory):
    def sho2(self):
        print(super().__a)


obj=Mumbai
obj.__a
```

```python
class Factory:
    __a='Pune'

    def __show(self):
        print('Hello i am in Pune factory')

obj=Factory
#obj.__a
obj.__show
```

```python
class Demo:
    def __init__(self):
        self.name='name'       #Public
        self._age=30           #Protected 
        self.__salary = 100000 #Private

    def show(self):
        print(self.name)
        print(self.age)
        print(self.__salary)




```
