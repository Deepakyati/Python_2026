```python
class Animal:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def __str__(self):
        return f"Hello how are you, my name is : {self.name}"
    
    def __add__(self, other):
        return f'sum of ages are : {self.age + other.age}'

obj = Animal('Lion',12)
obj1=Animal('dolphin',10)


print(obj + obj1)
```
