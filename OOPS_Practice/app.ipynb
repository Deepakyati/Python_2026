```python
#Constructor

class ABC:
    def __init__(self, a,b,c):
        self.a=a
        self.b=b
        self.c=c
```

```python
deepak = ABC(2,3,4)
print(deepak.c)
print(deepak.a)
print(deepak.b)
```

```python
xyz=ABC(10,20,30)

xyz.a
```

```python
#Types of attributes

class Animal:
    name='Lion' #class attribute

    def __init__(self,age):
        self.age=age #instance attribute

    def show(self): #instance method
        print(f'How are you? your age is {self.age}')

    @classmethod
    def hello(cls): #class method
        print(f'how r u broh {cls.name}')

    @staticmethod
    def static():
        print('how are youuuuuuuuuu')

obj = Animal(12)

obj.show()

obj.hello()

obj.static()

    

```
