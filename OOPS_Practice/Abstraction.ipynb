```python
from abc import ABC, abstractmethod

class abstract(ABC):
    @abstractmethod
    def perimeter(self):
        pass

    @abstractmethod
    def area(self):
        pass

class square(abstract):
    def __init__(self,side):
        self.side = side


class circle(abstract):
    def __init__(self,radius):
        self.radius=radius
    def perimeter(self):
        pass
    def area(self):
        pass

obj=circle(7)
```
