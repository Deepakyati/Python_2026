Distribution Plots

used for quantitative plots, like age, salary,....

  - distplot/histplot
  - jointplot
  - pairplot
  - rugplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
sns.distplot(df['total_bill'])


"""`distplot` is a deprecated function and will be removed in seaborn v0.14.0.

Please adapt your code to use either `displot` (a figure-level function with
similar flexibility) or `histplot` (an axes-level function for histograms).

For a guide to updating your code to use the new functions, please see
https://gist.github.com/mwaskom/de44147ed2974457ad6372750bbe5751

  sns.distplot(df['total_bill'])"""
```

```python
sns.histplot(df['total_bill'], bins=100)
```

```python
sns.histplot(df['total_bill'], bins=20) #bin size by default is 20
```

```python
sns.histplot(df['total_bill'], bins=20, kde=True) #kde for kernel density estimation
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'], bins=20, kde=True) 
plt.subplot(1,2,2)
sns.histplot(df['tip'], bins=20, kde=True) 
```

Joint Plot

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='scatter')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg')
```

Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='sex')
```

```python
sns.pairplot(df, hue='sex', palette='coolwarm')
```

```python
sns.pairplot(df, hue='sex', palette='rainbow')
```

```python
sns.pairplot(df, hue='size', palette='rainbow')
```

Rug Plot

```python
sns.rugplot(df)
```

```python
sns.rugplot(x='total_bill', y='tip', data=df)
```
