Matrix Plot

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
#correlation values between -1 and 1 
# 1 means perfect positive correlation
#-1 means perfect negative correlation
```

```python
tips_corr=tips[['total_bill', 'tip', 'size']]

tips_corr
```

```python
tips_corr.corr()
```

```python
sns.heatmap(tips_corr.corr())
```

```python
sns.heatmap(tips_corr.corr(), annot=True)
```

```python
sns.heatmap(tips_corr.corr(), annot=True, cmap='coolwarm')
```

Cluster Map

```python
sns.clustermap(tips_corr.corr())
```

```python
sns.clustermap(tips_corr.corr(), annot=True, cmap='coolwarm')
```
