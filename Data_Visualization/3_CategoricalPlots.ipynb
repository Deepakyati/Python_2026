Categorical Plots

  - boxplot
  - violinplot
  - stripplot
  - swarmplot
  - barplot
  - countplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.countplot(x=df['sex'])
```

```python
sns.countplot(x=df['sex'], hue=df['smoker']) #hue adds another categorical variable to the plot 
```

Bar Plot

```python
sns.barplot(x=df['sex'], y=df['total_bill'])
```

```python
sns.barplot(x=df['sex'], y=df['total_bill'], estimator=np.mean) #estimator can be changed to other functions like np.median, np.std etc. 
```

```python
sns.barplot(x=df['sex'], y=df['total_bill'], estimator=np.std)
```

```python
sns.barplot(x=df['sex'], y=df['total_bill'], estimator=np.sum)
```

```python
sns.barplot(x=df['sex'], y=df['tip'], estimator=np.sum)
```

Box Plot

```python
sns.boxplot(x=df['day'], y=df['total_bill'])
```

```python
sns.boxplot(x='tip', y='day', data=df)
```

```python
sns.boxplot(x='tip', y='day', data=df, palette='rainbow')
```

Violin Plot

boxplot and violinplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be â€œoutliersâ€ using a method that is a function of the inter-quartile range.

```python
sns.violinplot(x='day', y='total_bill', data=df) #violin plot combines aspects of boxplot and KDE plot
```

stripplot and swarmplot

The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they donâ€™t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).
