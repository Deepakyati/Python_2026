```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Matplotlib Basics

```python
x=np.linspace(1,10,10)
x
```

```python
y=x**2
y
```

```python
plt.plot(x,y)
```

```python
plt.title("Simple Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

SubbPlots in Matplotlib

```python
plt.subplot(2,2,4)#(rows, columns, position number)
plt.plot(x,y) 
plt.subplot(2,2,1)
plt.plot(x,y**2)
plt.subplot(2,2,3)
```

```python
plt.subplot(6,6,1)
plt.plot(x**2,y)
plt.subplot(6,6,6)
plt.plot(x,y**2)
```

Object Oriented way

```python
fig=plt.Figure()
fig
```

```python
axis1=fig.add_axes([0.1,0.1,0.5,0.5]) #left, bottom, width, height (range 0 to 1)
axis1.plot(x,y)
axis1.set_title("Big Plot")
axis1
fig
```

```python
fig=plt.Figure(figsize=(8,4),dpi=100)
axis2=fig.add_axes([0.1,0.1,0.8,0.8])
axis2.plot(x,y)
fig
```

Types of Plot

```python
plt.scatter(x,y)
```

```python
plt.hist(y)
```

```python
plt.boxplot(y)
```

Saving Plots

```python
fig=plt.figure()
ax=fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("myplot.png",dpi=200)
```
