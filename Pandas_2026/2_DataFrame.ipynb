DataFrames

```python
#combine multiple series = dataframe
import numpy as np
import pandas as pd
```

```python
#creating a dataframe

data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2 = pd.DataFrame(data_list)
df2
```

```python
columns = ["Name","Age","City","Salary"]
df2 = pd.DataFrame(data_list,columns =columns)
df2
```

```python
#selection and indexing of columns

df2['Name']  #single column selection
```

```python
df2[['Name','Age']]  #multiple column selection
```

```python
#create a new coloumn
#df2['Experience'] = [5,7,3,10]
df2['Designation'] = ["Manager","Developer","Analyst","Consultant"]
df2
```

```python
#removing a column
df2.drop('Designation',axis=1)
df2
```

```python
#removing a column
df2.drop('Designation',axis=1,inplace=True)
df2
```

```python
df2.drop(["City","Salary"],axis=1)
df2
```

```python
df2.drop(0,axis=0,inplace=True)
df2
```
