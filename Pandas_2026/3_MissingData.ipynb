Missing Data

```python
import numpy as np
import pandas as pd
```

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
#Removing missing values

df.dropna()  # Drops rows with any NaN values
```

```python
df
```

```python
#threshold can be specified to drop rows with less than certain non-NaN values

df.dropna(thresh=3)  # Keeps rows with at least 3 non-NaN values
```

```python
df
```

```python
#filling missing values

df.fillna(0)  # Fills NaN values with 0
```

```python
values={'A':0, 'B':100,'C':300,'D':400}
df.fillna(value=values)  # Fills NaN values with specified values per column
```

```python
values={'A': df['A'].mean(), 'C': df['C'].median(), 'D': 0}
df.fillna(value=values)  # Fills NaN in 'A' with mean of 'A', 'C' with median of 'C', 'D' with 0
```

```python
df.fillna(df.mean())  # Fills NaN values with the mean of each column   
```
