Numpy indexing and Slicing of Matrix

```python
#Numpy indexing and Slicing of Matrix

arr=np.arange(1,31).reshape(6,5)
arr
```

```python
arr[0] #print 1st row
```

```python
arr[1] #print 2nd row
```

```python
arr[2] #print 3rd row
```

```python
arr[0,0]
```

```python
arr[5,4]
```

```python
arr
```

```python
arr[0:2,1:3]
```

```python
arr[3:6,3:6]
```

```python
arr[3:,3:]
```

```python
#to access coloumn

arr[:,1:4]
```

```python
#Boolean Indexing

arr=np.arange(11,21)
arr
```

```python
bool_index=arr%2==0
bool_index
```

```python
arr=arr[bool_index] #print only True values
arr
```
