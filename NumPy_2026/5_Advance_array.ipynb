Advance array manipulation

```python
import numpy as np
```

```python
#array manipulation

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
```

```python
#stacking array

a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
```

```python
np.vstack((a,b))  #vertical stack
```

```python
np.hstack((a,b)) #horizontal stack
```

```python
np.column_stack((a,b))
```

```python
#Splitting Array

c=np.arange(16).reshape(4,4)
c
```

```python
np.hsplit(c,2)
```

```python
np.hsplit(c,3) # 4 elements can not be split in 3 equal part
```

```python
np.hsplit(c,4)
```

```python
np.vsplit(c,2)
```

```python
np.vsplit(c,4)
```
